```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df=pd.read_csv(r"/home/arun/Downloads/Salary_Data[1].csv")
df.dropna(inplace=True)
le1=LabelEncoder()
le2=LabelEncoder()
le=LabelEncoder()
df["Gender"]=le.fit_transform(df["Gender"])
df["Education Level"]=le1.fit_transform(df["Education Level"])
df["Job Title"]=le2.fit_transform(df["Job Title"])
df


                              

```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
x=df[["Job Title","Education Level",'Years of Experience']]
y=df['Salary']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,y_pred)

```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,y_pred)
r2_score(y_test,y_pred)

```
